```python
import zipfile
import os

def zip_folder(folder_path, output_zip_path):
    """
    יוצר קובץ ZIP מתיקייה
    
    :param folder_path: נתיב לתיקייה שרוצים לדחוס
    :param output_zip_path: נתיב לקובץ ZIP שייווצר
    """
    with zipfile.ZipFile(output_zip_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                file_path = os.path.join(root, file)
                # שומר את הנתיב היחסי בתוך הזיפ
                arcname = os.path.relpath(file_path, folder_path)
                zipf.write(file_path, arcname)

# דוגמה לשימוש
folder_to_zip = "videoshap"
output_file = "videoshap_paper.zip"
zip_folder(folder_to_zip, output_file)
```
